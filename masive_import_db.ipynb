```python
import pandas as pd
import MySQLdb
import datetime

# Configuración de la conexión a la base de datos
host = 'localhost'
usuario = 'root'
contraseña = 'root'
base_datos = 'spacebikes'

# Ruta del archivo CSV
archivo_csv = './raw_data/nuevo2.csv'

# Leer el archivo CSV usando pandas
df = pd.read_csv(archivo_csv)
df['precio'].fillna(0, inplace=True)

df
```

```python
# Establecer conexión a la base de datos MySQL
conexion = MySQLdb.connect(host=host, user=usuario, passwd=contraseña, db=base_datos)
cursor = conexion.cursor()

# Iterar sobre las filas del DataFrame e insertar en la base de datos
for indice, fila in df.iterrows():
    
    consulta = f"INSERT INTO space_producto (nombre, stock, precio, proveedor_id, id_prod_proveedor, descripcion, categoria) VALUES ('{fila['nombre']}', '{fila['stock']}', '{fila['precio']}', '{fila['proveedor_id']}', '{fila['id_prod_proveedor']}', '{fila['descripcion']}', '{fila['categoria']}');"
    
    # Ejecutar la consulta
    cursor.execute(consulta)

# Confirmar los cambios y cerrar la conexión
conexion.commit()
conexion.close()

print("Datos insertados correctamente en la base de datos.")
```
